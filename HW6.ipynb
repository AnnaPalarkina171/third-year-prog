```
from bs4 import BeautifulSoup 
from typing import List
import requests 
import urllib
import re

#Тут почти ничго нет, с курсовой не успеваю, 
#сдаю, чтобы потом была возможность переделать

def links_from_text(text):
    soup = BeautifulSoup(text)
    content = soup.find("div", {"id": "mw-content-text"})
    
    links = content.find_all("a")
    
    for link in links:
        href = link.get('href', '')
        if href.startswith('/wiki'): 
          href_uni = urllib.parse.unquote(href)
          if href.endswith('.svg') or href.endswith('.jpg') or href.endswith(
          '.png') or href_uni.startswith(
          '/wiki/Служебная:') or href_uni.startswith('/wiki/Шаблон:'):
            pass
          else:
            yield href 
            

start_url = 'https://ru.wikipedia.org/wiki/Теория струн'

for url in links_from_text(requests.get(start_url).text):
    print(urllib.parse.unquote(url))
```
